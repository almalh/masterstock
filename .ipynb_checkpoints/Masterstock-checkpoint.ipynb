```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
stock = yf.Ticker("CAE.TO")
#Fetch financials
income_stmnt = stock.financials
cash_flow = stock.cashflow
balance_sheet = stock.balance_sheet


#Extract Total Revenue (reverse order, convert to DataFrame and transpose)
total_revenue = income_stmnt.loc["Total Revenue"].iloc[::-1].to_frame().transpose()

#Generate YoY increase series with same Timestamps as revenue
yoy_increase = total_revenue.iloc[0].copy()
yoy_increase.iloc[0] = np.nan
for n in range(total_revenue.size-1):
    yoy_increase.iloc[n + 1] = (total_revenue.iloc[0, n + 1] / total_revenue.iloc[0, n] - 1) * 100
    
#Fetch Operating Cash Flow (OCF) and Capital Expenditures (Capex)
ocf = cash_flow.loc["Total Cash From Operating Activities"]
capex = cash_flow.loc["Capital Expenditures"]

#Generate Free Cash Flow (FCF)
fcf = ocf.add(capex)

#Calculate FCF as a percentage of revenue
fcf_perc = fcf.divide(total_revenue.iloc[0]).multiply(100)   
    
    
```

```python
table = total_revenue.copy()
table.loc['YoY Increase (%)'] = yoy_increase
table.loc['Operating Cash Flow'] = ocf
table.loc['Capital Expenditures'] = capex
table.loc['Free Cash Flow'] = fcf
table.loc['% of Revenue'] = fcf_perc
table
```
